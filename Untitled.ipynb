```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Collect Data
data = {
    'Date': ['1/7/2024', '2/7/2024', '3/7/2024', '4/7/2024', '5/7/2024', '6/7/2024'],
    'Name': ['Sanjaay']*6,
    'Login': ['9:00 AM']*6,
    'Task': ['Other ATS Website Testing and R&D', 'Reference and documentation', 'Reference and document', 
             'Reference and document', 'Reference and document', 'Reference and document'],
    'Percentage': [20, 30, 50, 80, 100, 100],
    'Logout': ['1:30 PM', '2:00 PM', '4:00 PM', '2:00 PM', '2:00 PM', '2:00 PM']  # Corrected logout times
}

df = pd.DataFrame(data)

# Step 2: Calculate Total Hours Worked
def calculate_hours(logout_time):
    login_time = pd.to_datetime('9:00 AM', format='%I:%M %p')
    logout_time = pd.to_datetime(logout_time, format='%I:%M %p')
    
    # Calculate hours worked
    hours_worked = (logout_time - login_time).seconds / 3600
    
    return hours_worked

df['Hours Worked'] = df['Logout'].apply(calculate_hours)

# Step 3: Analyze Data
average_percentage = df['Percentage'].mean()
total_hours = df['Hours Worked'].sum()

print(f"Average Task Completion Percentage: {average_percentage}%")
print(f"Total Hours Worked: {total_hours} hours")

# Step 4: Visualize Data
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Date'], df['Percentage'], marker='o')
plt.title('Task Completion Percentage Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage')

plt.subplot(1, 2, 2)
plt.plot(df['Date'], df['Hours Worked'], marker='o', color='r')
plt.title('Hours Worked Over Time')
plt.xlabel('Date')
plt.ylabel('Hours Worked')

plt.tight_layout()
plt.show()

# Step 5: Generate Reports
# This step can involve using a library like ReportLab to create PDF reports.
```
